## Graphs

### Importing MatplotLib

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Have x and y values

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

### Plot y against x
### Show the plot

```python
plt.plot(x,y)
plt.show()
```

### Add a title for x
### Add a title for the plot
### Add labelsto x and y for the plot

```python
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

### Change the line: style, colour and width

```python
plt.plot(x,y, color = "red", linestyle= ":", linewidth = 4)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Add a second line & plot it

```python
y2 = 3*x
y2
```

```python
plt.plot(x,y, color = "red", linestyle= ":", linewidth = 4)
plt.plot(x,y2, color = "g", linestyle= "-", linewidth = 2)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Add a grid

```python
plt.plot(x,y, color = "red", linestyle= ":", linewidth = 4)
plt.plot(x,y2, color = "g", linestyle= "-", linewidth = 2)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

### Adding Sub-plots

```python
# .subplot(nrows, ncols, index)

#nrows: The number of rows in the grid of subplots.
#ncols: The number of columns in the grid of subplots.
#index: The index of the current subplot. The subplots are numbered from left to right and top to bottom starting from 1.

# add a 3rd line
y3 = 4*x


plt.subplot(1,3,1)
plt.plot(x,y, color = "red", linestyle= ":", linewidth = 4)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(x,y2, color = "g", linestyle= "-", linewidth = 2)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(1,3,3)
plt.plot(x,y3, color = "blue", linestyle= "-", linewidth = 3)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

#spaces the graphs
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, hspace=0.3, wspace=0.3)


plt.show()
```

## Bar charts

### Make a dictionary of students with names as keys and their score % for a test as their values

```python
student = {"Ismail": 88, "Anas": 90, "Josh": 65}
#put the keys in a list 
names = list(student.keys())
#put the values in a list
scores = list(student.values())

plt.bar(names,scores)
plt.title("My Bar chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

### Make a horizontal bar plot

```python
plt.barh(names,scores,color = "g")
# barh means bar hedge
plt.title("My Bar chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

## Scatter plots

```python
x = [10,20,30,40,50,60]
y = [10,2,9,21,5,6]

plt.scatter(x,y, color = "purple", s = 150)
plt.show()
```

### Changing this size and shape of the markers

```python
x = [10,20,30,40,50,60]
y = [10,2,9,21,5,6]

plt.scatter(x,y,marker = "x", color = "green", s = 70)
plt.show()
```

### Adding another set of scatter data

```python
x = [10,20,30,40,50,60]
y = [10,2,9,21,5,6]
y2 = [10,1,15,13,18,8]
# there are different types of markers e.g o x *
plt.scatter(x,y2,marker = "*", color = "purple", s = 90)
plt.scatter(x,y,marker = "x", color = "green", s = 70)
plt.show()
```

### Adding the scatter data as subplots

```python

plt.subplot(2,1,1)
plt.scatter(x,y,marker = "*", color = "purple", s = 90)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(2,1,2)
plt.scatter(x,y,marker = "x", color = "green", s = 70)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```

### Creating a boxplot

```python
# 25% value is 25% of the way into the list 75% value is 75% of the way into the list
one = [1,2,3,4,5,6,7,8,9] #smallest value is 1 largest is 9 median is 5 25% value is 3 75% value is 7
two = [1,2,3,4,5,1,2,3,4]
three = [4,8,2,9,10,3,6,7]
data = list([one,two,three])

plt.boxplot(data)
plt.show
```

### Pie Chart

```python
fruit = ["Banana","Satsuma","Pineapple","Apple","Orange"]
Quantity = [30,5,73,29,67]
plt.pie(Quantity,labels = fruit)
plt.show()
```

### Changing pie chart colours and adding percentages

```python
plt.pie(Quantity,labels = fruit,autopct = "%0.1f%%",
colors = ["yellow","pink","green","red","orange" ])
plt.show()
```

```python
plt.pie(Quantity,labels = fruit,autopct = "%0f%%",
colors = ["yellow","pink","brown","red","orange" ])
plt.show()
```

```python
plt.pie(Quantity,labels = fruit,autopct = "%0.0f%%",
colors = ["yellow","pink","brown","red","orange" ])
plt.show()
```

### Doughnut Chart

```python
plt.pie(Quantity,labels = fruit,autopct = "%0.0f%%",
colors = ["yellow","pink","brown","red","orange" ], radius = 2)
plt.pie([1],colors = ["w"], radius = 1)
# we essentially plot two pie charts inside each other play with the radi

#The [1] parameter passed to plt.pie() creates a small, white circle in the center of the pie chart. 
#This is a commonly used technique to create a "donut" chart, where the inner circle represents additional information or context.
```
